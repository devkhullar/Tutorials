## Bivariate plot with multiple elements
I am first making the scatterplot presented [here](https://seaborn.pydata.org/examples/layered_bivariate_plot.html)

```python
# Simulate data from a bivariate Gaussian
n = 10000
mean = [0, 0]
cov = [(2, .4), (.4, .2)]
rng = np.random.RandomState(0)
x, y = rng.multivariate_normal(mean, cov, n).T
```

```python
f, ax = plt.subplots(figsize=(6, 6))
sns.scatterplot(x=x, y=y, s=5, color=".15")
```

```python
f, ax = plt.subplots(figsize=(6, 6))
sns.scatterplot(x=x, y=y, s=5, color=".15")
sns.histplot(x=x, y=y, bins=50, pthresh=.1, cmap="mako")
```

```python
# Draw a combo histogram and scatterplot with density contours
f, ax = plt.subplots(figsize=(6, 6))
sns.scatterplot(x=x, y=y, s=5, color=".15")
sns.histplot(x=x, y=y, bins=50, pthresh=.1, cmap="mako")
sns.kdeplot(x=x, y=y, levels=5, color="w", linewidths=1)
```

This looks cool!

## Hexbin plot with marginal distributions

```python
sns.set_theme(style="ticks")
rs = np.random.RandomState(11)
x = rs.gamma(2, size=1000)
y = -.5 * x + rs.normal(size=1000)

sns.jointplot(x=x, y=y, kind="hex", color="#4CB391")
```

## [Controlling figure aesthetics](https://seaborn.pydata.org/tutorial/aesthetics.html)

```python
def sinplot(n=10, flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, n + 1):
        plt.plot(x, np.sin(x + i * 0.5) * (n + 2 - i) * flip)
```

```python
sinplot()
```

```python
sns.set_theme()
```

```python
sinplot()
```

```python
sns.set_style('dark')
sinplot()
```

```python
sns.set_style('white')
sinplot()
```

```python
sns.set_style("ticks")
sinplot()
```

```python
sinplot()
sns.despine() # to remove the spines
```

### Temporarily setting figure style 
This is kinda smart

```python
f = plt.figure(figsize=(6, 6))
gs = f.add_gridspec(2, 2)

with sns.axes_style('darkgrid'):
    ax = f.add_subplot(gs[0, 0])
    sinplot(6)

with sns.axes_style('white'):
    ax = f.add_subplot(gs[0, 1])
    sinplot(6)

with sns.axes_style('ticks'):
    ax = f.add_subplot(gs[1, 0])
    sinplot(6)

with sns.axes_style('whitegrid'):
    ax = f.add_subplot(gs[1, 1])
    sinplot(6)

    f.tight_layout()
```
